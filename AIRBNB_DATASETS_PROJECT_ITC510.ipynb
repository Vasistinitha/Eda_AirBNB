# Merging and EDA on AirBnb Final Dataset

```python
# Importing all the libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# loading the .csv files into the dataframe
df1 = pd.read_csv("Newyork_data1.csv")
df2 = pd.read_csv("Newyork_data2.csv")
```

```python
# merging 2 different datasets using inner join
cols_to_use = df2.columns.difference(df1.columns)
dfNew = pd.merge(df1, df2[cols_to_use], left_index=True, right_index=True, how='inner')
dfNew
```

```python
# droping an unwanted column from the dataframe
dfNew.drop(["Unnamed: 0"], axis = 1, inplace = True)
dfNew
```

```python
# fetching the column info 
dfNew.info()
```

```python
dfNew.describe()
```

```python
# plotting correlation heatmap
plt.figure(figsize=(12,6))
corr = dfNew.corr()
mask = np.triu(corr)
sns.heatmap(corr, annot= True, mask= mask, fmt = '.2f')
plt.title("correlation between attributes")
plt.show()
```

```python
# retriving the most expensive airbnb in the newyork city
test_df = dfNew.sort_values("price",ascending=False).head()[["name", "price"]]
test_df
```

```python
# retriving the most expensive airbnb in the newyork city
test_df=dfNew.sort_values("price",ascending=True).head()[["name", "price"]]
test_df
```

```python
# retriving the least busiest hotels in newyork city
test_df=dfNew.sort_values("availability_365",ascending=False).head()[["name", "availability_365"]]
test_df
```

```python
# plotting for differences in prices per review rate number
figure = plt.figure(figsize = (10,8))
ax = sns.lineplot(x = "review rate number", y = "price", data = dfNew)
plt.title("Differences in Prices Per Review Rate Number")
plt.show()
```

```python
# plotting Price variations for different Room Types
figure = plt.figure(figsize = (10,5))
ax = sns.barplot(y = "price", x = "room type", data = dfNew, color = '#4287f5')
plt.title("Price Variations for Different Room Types")
plt.show()
```

```python
# plotting pricing range for recent & old constructed Airbnb's
figure = plt.figure(figsize = (10,8))
ax = sns.lineplot(x = "Construction year", y = "price", data = dfNew)
plt.title("Pricing Range for Recent & Old Constructed Airbnb's")
plt.show()
```

```python
# plotting pricing range for recent & old constructed Airbnb's
figure = plt.figure(figsize = (10,8))
ax = sns.lineplot(x = "review rate number", y = "availability_365", data = dfNew)
plt.title("Availability against Review Rate Number")
plt.show()
```
