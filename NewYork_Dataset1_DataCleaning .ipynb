```python
# Importing pandas library
import pandas as pd
import datetime as dt
# Loading .Csv file into dataframe
df = pd.read_csv("listings_NY.csv")

# Returns first 5 rows of the dataframe
df.head(5)
```

```python
# Detect missing values in dataframe and returns boolean values
df.isnull()
```

```python
# Counts the missing values per each column in the dataframe.
df.isna().sum()
```

```python
# Drop unwanted columns from the table.
print(df.drop(["license","latitude","longitude","number_of_reviews_ltm","reviews_per_month","price", "room_type","neighbourhood_group", "neighbourhood"],axis = 1,inplace = True))
```

```python
# Counts the missing values per each column in the dataframe.
df.isna().sum()
```

```python
# Returns the size of the table.
df.shape
```

```python
#Keep the DataFrame with valid entries in the same variable.
new = df.dropna(inplace = True)
print(new)
```

```python
# Summing up all duplicate rows
print(df.duplicated().sum()) 

# To remove duplicate rows based on all columns.
print(df.drop_duplicates(inplace = True))

# To remove duplicates on specific column(s)
print(df.drop_duplicates(subset = ['name', 'host_id', 'host_name'], inplace = True))
print(df.drop_duplicates(subset = ['host_id', 'host_name'], inplace = True))
print(df.drop_duplicates(subset = ['name'], inplace = True)) 
print(df.drop_duplicates(subset = ['host_id'], inplace = True))

# Returns the size of the table
df.shape
```

```python
# Insert and add column into DataFrame at specified location.
new = df.insert(3, "City", "New York")
print(new)

#changing the datetime of last_review to mm/dd/yyyy
df["last_review"] = pd.to_datetime(df["last_review"])

#the string (object) format changes the format to mm//dd//yyyy
df["last_review"] = pd.to_datetime(df["last_review"]).dt.strftime('%m-%d-%Y')
df
```
