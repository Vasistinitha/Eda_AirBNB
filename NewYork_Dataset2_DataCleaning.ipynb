```python
#Importing required libraries
import pandas as pd
import datetime as dt

# Loading .Csv file into dataframe
df = pd.read_csv("Airbnb_Open_Data.csv")

# Returns first 5 rows of the dataframe
df.head()
```

```python
df.info()
```

```python
#counts the missing values per each column in the dataframe.
df.isna().sum()
```

```python
# Drop unwanted columns from the table.
df.drop(["license","house_rules","reviews per month","lat","long","neighbourhood group", "neighbourhood", "country code", "host_identity_verified"],axis = 1,inplace = True)
df
```

```python
#Keep the DataFrame with valid entries in the same variable.
df.dropna(inplace=True)

# To remove duplicate rows based on all columns.
df.drop_duplicates(inplace=True)

# To remove duplicates on specific column(s)
df.drop_duplicates(subset = ['NAME', 'host id', 'host name'], inplace = True)
df.drop_duplicates(subset = ['host id', 'host name'], inplace = True)
df.drop_duplicates(subset = ['NAME'], inplace = True)
df.drop_duplicates(subset = ['host id'], inplace = True)
df
```

```python
#changing the datetime of last_review to mm/dd/yyyy
df["last review"] = pd.to_datetime(df["last review"])

#the string (object) format changes the format to mm//dd//yyyy
df["last review"] = pd.to_datetime(df["last review"]).dt.strftime('%m-%d-%Y')
df
```

```python
# removing special character ($) from price and service fee columns
df['price'] = df['price'].str.replace('\W', '', regex=True)
df['service fee'] = df['service fee'].str.replace('\W', '', regex=True)
df
```

```python
# Renaming specific label names in dataframe
df.rename({"service fee" : "service_fee", "NAME" : "name", "host id" : "host_id", "host name" : "host_name", "minimum nights":"minimum_nights", "number of reviews":"number_of_reviews","last review": "last_review", "calculated host listings count":"calculated_host_listings_count","availability 365":"availability_365"}, axis = 1, inplace = True)
df
```

```python
# altering the price and service fee charges from int to float 
df.price = df.price.astype("float64")
df.service_fee = df.service_fee.astype("float64")
df
```

```python
#counts the missing values per each column in the dataframe.
df.isna().sum()
```

```python
# Write DataFrame to an CSV file
df.to_csv("Newyork_data2.csv")
```
